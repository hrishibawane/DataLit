```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

x_test = pd.read_csv('X_test.csv')
y_train = pd.read_csv('Y_train.csv')
y_test = pd.read_csv('Y_test.csv')

x_train = dfx[['ApplicantIncome','LoanAmount','CoapplicantIncome','Loan_Amount_Term','Credit_History']]
x_test = x_test[['ApplicantIncome','LoanAmount','CoapplicantIncome','Loan_Amount_Term','Credit_History']]

y_train = y_train['Target']
y_test = y_test['Target']

```

```python
model_naive = GaussianNB()
model_naive.fit(x_train, y_train)
```

```python
y_pred = model_naive.predict(x_test)
y_pred

accuracy_score(y_test, y_pred)
```

```python
e=0
u=0
for a,b in zip(y_test,y_pred):
    if a==b:
        e+=1
    else:
        u+=1
    
e,u
```
